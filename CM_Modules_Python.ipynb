## Numpy

Comme son nom l'indique, c'est la bibliothèque qui gère les outils de calcul numérique scientifique.

L'objet phare de Numpy est le *ndarray* i.e. le *n-dimensional array*.

Un *ndarray* permet souvent un traitement plus rapide des données qu'une liste Python.

Une différence notable cependant : les éléments d'un *ndarray* doivent être de même type.

```python
import numpy as np

a = np.array([[10, 2,3], [4,5,6]])
```

```python
a
```

Une des caractéristiques les plus importantes des *ndarray* est leur aspect **vectoriel** :

```python
3*a
```

```python
a + a 
```

```python
a * a
```

## Matplotlib

C'est le module le plus célèbre pour tracer des graphiques, en particulier le sous-module `pyplot`

```python
import matplotlib.pyplot as plt
import numpy as np
```

On veut représenter la fonction $t \mapsto 1 +\sin(2\pi t)$ sur l'intervalle $[0,2]$

```python
t = np.arange(0.0, 2.0, 0.01) # le temps 
v = 1 + np.sin(2 * np.pi * t) # une fonction de t
```

```python
fig, ax = plt.subplots()
ax.plot(t, v)

ax.set(xlabel='temps (s)', ylabel='voltage (mV)',
       title='Trop beau mon premier dessin')
ax.grid()
```

```python
Z = np.random.rand(20, 30)

fig, (ax0, ax1) = plt.subplots(2, 1)

ax0.pcolor(Z)
ax1.pcolor(Z, edgecolors='k', linewidths=.5)
```

## Sympy

Module de calculsymbolique qui pourrait permettre de vous rendre des services...

```python
import sympy as sy
```

```python
k,n = sy.symbols('k n')
sy.init_printing() # pour un joli affichage
sy.summation(3**k, (k,1,n))
```

```python
f = lambda x: (x**2 + x)**3
```

```python
x = sy.symbols('x')
f(x)
```

```python
sy.diff(f(x), x)
```

```python
g = lambda x : (sy.exp(x) + 1)/(sy.exp(x) - 1)
g(x)
```

```python
sy.limit(g(x), x, -sy.oo)
```

```python
sy.diff(g(x), x)
```

```python
sy.factor(sy.diff(g(x), x))
```

```python
sy.plot(g(x),(x,1,10))
```

```python
sy.solve(x**2 + x - 1,x)
```

```python
sy.integrate(sy.log(x), x)
```

## Un exemple de GUI : Tkinter

```python
import tkinter as tk
from tkinter import ttk as ttk
from tkinter.messagebox import showerror
```

```python
fenetre = tk.Tk() # on crée une instance
fenetre.geometry('100x100')

def Bonjour() -> None:
    msg = tk.messagebox.showinfo("Bonjour Angers", "Bonjour les L3")
    
B = ttk.Button(fenetre, text='Salut', command=Bonjour)
B.grid(row=4, column=2)

def _quit():
    fenetre.quit()     # stops mainloop
    fenetre.destroy()  # this is necessary on Windows to prevent
                    # Fatal Python Error: PyEval_RestoreThread: NULL tstat
        
        

buttonq = ttk.Button(fenetre, text='Cliquez ici pour tout fermer', command=_quit)
buttonq.grid(column=2,row=9)

fenetre.mainloop()
```
